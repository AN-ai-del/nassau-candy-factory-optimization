```python
duplicates = df.duplicated().sum()

print(f"Duplicate Rows : {duplicates}")
```

### Observation

Duplicate rows are evaluated because repeated transactions may bias analytical results and machine learning models.

Only true duplicate records should be removed.

```python
duplicate_orders = (
    df["Order ID"]
    .duplicated()
    .sum()
)

print(duplicate_orders)
```

```python
df["Order ID"].value_counts().head(10)
```

### Observation

Multiple records per Order ID are expected because a single customer order may contain multiple products.

Therefore, repeated Order IDs do not necessarily indicate duplicate transactions.

```python
print(f"Negative Sales Records : {(df['Sales'] < 0).sum()}")
print(f"Negative Cost Records  : {(df['Cost'] < 0).sum()}")
print(f"Negative Profit Records: {(df['Gross Profit'] < 0).sum()}")
print(f"Zero Unit Orders       : {(df['Units'] == 0).sum()}")
```

### Observation

Financial variables are validated to ensure that no impossible business transactions exist.

Negative sales, zero-unit orders, or invalid costs may indicate data quality issues requiring further investigation.

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], format="mixed")

df["Ship Date"] = pd.to_datetime(df["Ship Date"], format="mixed")
```

```python
df[["Order Date","Ship Date"]].dtypes
```

### Observation

Order Date and Ship Date have been converted into datetime format.

This enables time-based analysis, feature engineering, and shipping lead-time calculations.

```python
df["Shipping Lead Time"] = (
    df["Ship Date"] -  df["Order Date"]
).dt.days
```

```python
df["Shipping Lead Time"].describe()
```

```python
df[["Order Date",
    "Ship Date",
    "Shipping Lead Time" 
]].head()
```

```python
invalid_dates = (
    df["Shipping Lead Time"] < 0
).sum()

print(f"Invalid Shipping Records : {invalid_dates}")
```

### Observation

Shipping lead time must always be zero or greater.

Negative values indicate invalid business records that require investigation.
